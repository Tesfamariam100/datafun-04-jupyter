```python
# Load the Iris dataset using Seaborn
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

##Inspect histogram by numerical column - colum petal_width

df['petal_width'].hist()

plt.title('Figure1: ' + 'Histogram For ' + 'Petal Width')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Compute statistics
mean_petal_width = df['petal_width'].mean()
min_petal_width = df['petal_width'].min()
max_petal_width = df['petal_width'].max()
range_petal_width = max_petal_width - min_petal_width
# Print statistics
print("Mean Petal Width:", mean_petal_width)
print("Range of Petal Width:", range_petal_width)
```

## Observation of Figure1

- Central Tendency: Between 1.0 and 2.0, it suggests a normal distribution.
- Outliers: Petal widths above 3.0 or below 0.5, represent outliers.
- Data Spread: Indicates greater variability in petal width measurements.( from 0.5 to 3.5) .
Finally, the statistics are printed along with the histogram plot.






```python
import seaborn as sns

# Create a violin plot for petal width
sns.violinplot(x='species', y='petal_width', data=df)

# Set the title
plt.title('Violin Plot for Petal Width by Species')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot with species on the x-axis and petal width on the y-axis
sns.boxplot(x='species', y='petal_width', data=df, hue='species', palette='pastel', legend=False)

# Add title and labels
plt.title('Boxplot of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the pairplot with hue='species'
pairplot = sns.pairplot(df, hue='species')

# Adjust the size of the pairplot
pairplot.fig.set_size_inches(8, 8)  # Set the figure size (width, height) in inches

# Add title
plt.suptitle('Pairplot of Iris Dataset by Species', y=1.02)  # Adjust the y-coordinate for title positioning

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram with bars of different colors for each category (species)
sns.histplot(data=df, x='petal_width', hue='species', multiple='stack', palette='colorblind')

# Add title and labels
plt.title('Histogram of Petal Width by Species')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```


# Conclusion
### Central Tendency: 
The data appears to be concentrated around the range of 1.0 to 2.0, indicating a normal distribution tendency.
### Outliers: 
Petal widths falling outside the range of 0.5 to 3.0 could be considered as outliers due to their significant deviation from the main distribution.
### Data Spread: 
The spread of the data suggests variability in petal width measurements, spanning from 0.5 to 3.5, indicating diverse characteristics within the dataset.
